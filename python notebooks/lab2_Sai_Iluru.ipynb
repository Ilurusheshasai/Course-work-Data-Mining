### Creating a data profile

```python
import pandas as pd
```

```python
#loading the dataset / reading CSV file
df = pd.read_csv(r"C:\Users\Sai_iluru\OneDrive - UMBC\Desktop\UMBC\733\python notebooks\breast_cancer.csv")
df.head()
```

```python
import pandas as pd


# Create an empty DataFrame for the data profile
profile_dfs = []

# Iterate through each column and extract information
for col in df.columns:
    data_type = 'int64' if pd.api.types.is_numeric_dtype(df[col]) else 'Object'
    
    if pd.api.types.is_numeric_dtype(df[col]):
        unique_values = df[col].nunique()
        min_val = df[col].min()
        max_val = df[col].max()
    else:
        unique_values = df[col].nunique()
        min_val, max_val = 'NaN','NaN'
    
    profile_col_df = pd.DataFrame({
        'Column Name': [col],
        'Data Type': [data_type],
        'Unique Values': [unique_values],
        'Min': [min_val],
        'Max': [max_val]
    })
    
    profile_dfs.append(profile_col_df)

# Concatenate all DataFrames into a single data profile DataFrame
profile_df = pd.concat(profile_dfs, ignore_index=True)

# Print the data profile table
print(profile_df)
```
