```python
#Create a bar plot of mean salary estimate from the same top 5 industries - see sample output #2: Hint
#You will need to first clean the “Salary Estimate” column
# Remove rows with -1 in any column
grouping_elements = list(top_5_industries.index)
#print(grouping_elements)
task3_df = df[['Industry', 'Salary Estimate']]
task3_df.replace('-1', np.nan, inplace=True)
# Drop rows containing NaN
task3_df.dropna(inplace=True)
task3_df
```

```python
import re
# Define a custom function that operates on each value
def cal_avg_sal(salary_range):
    # Split the range into two values
    start_salary, end_salary = salary_range.split('-')
    
    # Remove 'k' from each value and convert to integers
    start_salary = re.sub(r'[^0-9]', '', start_salary)
    end_salary = re.sub(r'[^0-9]', '', end_salary)
    
    # Calculate the average salary
    avg_salary = (int(start_salary) + int(end_salary))/2
    return avg_salary
# Apply the custom_function to the 'Column1' column
#task3_df['Avg_salary'] = task3_df['Salary Estimate'].apply(cal_avg_sal)
task3_df['Mid Salary'] = task3_df['Salary Estimate'].apply(cal_avg_sal)
task3_df
```

```python
new_df_task_3 = task3_df[task3_df['Industry'].isin(['IT Services', 'Staffing & Outsourcing', 'Health Care Services & Hospitals', 'Computer Hardware & Software', 'Investment Banking & Asset Management'])]
new_df_task_3
```

```python
result_df = new_df_task_3.groupby('Industry')['Mid Salary'].mean().reset_index()
result_df
```

```python

# Plot the graph industry vs Average Salary
result_df.plot(x='Industry', y='Mid Salary', kind='bar', figsize=(12, 6))

# Add labels and a title
plt.xlabel('Industries')
plt.ylabel('Average salaries')
plt.title('Bar plot of Average salaries from top 5 industries')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Representing Rating of industries at different locations do determine best locations to work  

task3_sub_task = df[['Location', 'Rating']]
rating_df = task3_sub_task.groupby('Location')['Rating'].mean().reset_index()

# Define custom colors based on value ranges
colors = []

for value in rating_df['Rating']:
    if value < 3:
        colors.append('red')
    elif value >= 3 and value < 4:
        colors.append('orange')
    else:
        colors.append('green')

# Plot the Location vs avg Rating of comapnies
rating_df.head(15).plot(x='Location', y='Rating', kind='bar', figsize=(12, 6),color=colors)

# Create a legend
legend_labels = ['Rating < 3', ' Rating between 3-4', ' Rating >= 4']
legend_colors = ['red', 'orange', 'green']
legend_handles = [plt.Line2D([0], [0], color=color, lw=4) for color in legend_colors]

plt.legend(legend_handles, legend_labels, title='Representaion of ratings')

# Add labels and a title
plt.xlabel('Location')
plt.ylabel('Average Rating of companies')
plt.title('Bar plot of cities with top rated companies to work')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
